```python
from bs4 import BeautifulSoup as bs
import requests as rq

```

```python
url = 'https://quotes.toscrape.com/'
```

```python
req = rq.get(url)
```

```python

bs1 = bs(req.text,"html")
print(bs1.prettify())
```

```python
text = list(bs1.stripped_strings)
```

```python
text
```

```python
print(text)
```

```python
len(text)
```

```python
data1 = bs1.find_all('span',class_ = "text",itemprop_ = "")
```

```python
data1
```

```python
for texts in data1:
    print(texts.text)
```

```python
resp = bs1.find_all('large')
```

```python
len(resp)
```

```python
resp = bs1.find_all('small')
```

```python
len(resp)
```

```python
for name in resp:
    print(name.text)
```
